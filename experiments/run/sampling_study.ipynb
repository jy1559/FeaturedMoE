# Sampling Study

KuaiRec / lastfm downsampling strategies for dev-scale FMoE experiments.

This notebook delegates the heavy lifting to `experiments/tools/analyze_sampling_strategies.py` and then reads the generated reports.

```python
from pathlib import Path

repo_root = Path('/workspace/jy1559/FMoE')
script_path = repo_root / 'experiments' / 'tools' / 'analyze_sampling_strategies.py'
output_dir = repo_root / 'experiments' / 'run' / 'artifacts' / 'analysis' / 'sampling_study'
datasets = 'KuaiRec,lastfm'
ratios = '0.03,0.05'
seed = '42'

script_path, output_dir
```

```python
import subprocess

subprocess.run(
    [
        'python3',
        str(script_path),
        '--datasets', datasets,
        '--ratios', ratios,
        '--seed', seed,
        '--output-dir', str(output_dir),
    ],
    check=True,
)
```

```python
for dataset in ('KuaiRec', 'lastfm'):
    report = output_dir / f'{dataset}_sampling_study.md'
    print(report.read_text())
    print('\n' + '=' * 100 + '\n')
```
